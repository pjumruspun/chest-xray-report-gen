```python
import pandas as pd
import numpy as np
```

```python
def formatter(s, split, seed, project_dim):
    line_number = 0
    ls = []
    conditions = []
    for l in s.split('\n'):
        l = l.strip()
        if len(l) > 0:
            l = l.split('\t')
            if line_number == 0:
                columns = l
            else:
                conditions.append(l[0])
                ls.append(l[1:])
            line_number += 1

    arr = np.array(ls)
    df = pd.DataFrame(arr)
    df.index = pd.Index(conditions, name=columns[0])
    df.columns = pd.Index(columns[1:])
    df.to_csv(f'results/{split}_results_{project_dim}_{seed}.csv', index=False)
    return df
```
